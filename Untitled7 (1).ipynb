```python
import time
import random

def game():
    
    # 游戏介绍
    print("欢迎来到骰子游戏!")
    print("\n游戏规则：")
    print("- 你开始时拥有100积分。")
    print("- 游戏使用3个骰子，每个骰子的点数为1-6。")
    print("- 投掷三个骰子后，他们的点数范围是3～18。")
    print("- 将3～10划分为“小”，11～18划分为“大”。")
    print("- 你可以选择“大”或“小”，“单”或“双”，或特定的点数（3-18）来进行投注。")
    print("- 大小和单双的奖金倍率为1.95倍，特定点数的奖金倍率随点数而变。")
    print("\n倍率详情：")
    for k, v in multiplier_dict.items():
        print(f"{k}点：{v}倍")
    print("\n开始游戏!")
    
    points = 100  # 初始积分
    while points > 0:
        print("\n当前的积分为:", points)

        # 获取用户投注选择和金额
        size_choice = input('请输入 "大" 或 "小" (若不投此项，请留空)：')
        size_bet = float(input("请输入您的投注金额: ")) if size_choice else 0
        
        parity_choice = input('请输入 "单" 或 "双" (若不投此项，请留空)：')
        parity_bet = float(input("请输入您的投注金额: ")) if parity_choice else 0
        
        number_choice = input("选择一个点数 (3-18) (若不投此项，请留空)：")
        number_bet = float(input("请输入您的投注金额: ")) if number_choice else 0

        # 投掷骰子并获取结果
        dice1, dice2, dice3 = random.randint(1, 6), random.randint(1, 6), random.randint(1, 6)
        total = dice1 + dice2 + dice3

        # 让结果在5秒后显示
        print("骰子正在滚动...")
        time.sleep(5)
        print(f"投掷的点数和为: {total}, 点数为: {dice1}, {dice2}, {dice3}")

        # 结果判断并计算积分
        if size_choice:
            if size_choice == "大" and total > 10:
                points += size_bet * 1.95
            elif size_choice == "小" and total <= 10:
                points += size_bet * 1.95
            else:
                points -= size_bet

        if parity_choice:
            if parity_choice == "单" and total % 2 != 0:
                points += parity_bet * 1.95
            elif parity_choice == "双" and total % 2 == 0:
                points += parity_bet * 1.95
            else:
                points -= parity_bet

        if number_choice:
            if int(number_choice) == total:
                points += number_bet * multiplier_dict[int(number_choice)]
            else:
                points -= number_bet

        if points < 1:
            print("请v THT 100米以继续游戏")
            break

multiplier_dict = {
    3: 240, 4: 210, 5: 180, 6: 150, 7: 120, 8: 90, 9: 60, 10: 30,
    11: 30, 12: 60, 13: 90, 14: 120, 15: 150, 16: 180, 17: 210, 18: 240
}

game()
```
